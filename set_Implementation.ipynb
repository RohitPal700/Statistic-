```python
#set operation >> union , intersection ,difference, superset,subset,symmetric difference
```

```python
set1={34,67,78,100}
set2={67,78}
```

```python
set1 & set2 # Intersection 
```

```python
set1.intersection(set2) # MEthod 2 find the intersection 
```

```python
set1|set2 # union 
```

```python
set1.union(set2)   # Method 2 find the union 
```

```python
set1 - set2 # Difference 
```

```python
set1.difference(set2)  # Method 2 find the difference
```

```python
set1.issuperset(set2) # Here we check set1 is superset of set2 
```

```python
set1.issubset(set2) ## Here we check set1 is subset of set2 
```

```python
set2.issubset(set1) ## Here we check set2 is subset of set1
```

```python
#symmetric difference
set1.symmetric_difference(set2)
```

```python
#method 2 find the symmetric difference 
set1^set2
```
